# Import Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

# Import dataset

```python
# Load dataset
data = pd.read_csv('glass.csv')

# Pisahkan antara atribut dan label
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

# Splitting data into the training test dan test set

```python
# Split dataset menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Scalling Feature

```python
# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# melatih model klasifikasi KNN pada Training set

```python
from sklearn.neighbors import KNeighborsClassifier
# Training KNN pada set pelatihan
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, y_train)
```

# Prediksi

```python
# Prediksi hasil pada set pengujian
y_pred = classifier.predict(X_test)
```

# Confusion Matrix

```python
# Membuat confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Visualisasi

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Buat ListedColormap untuk plot
colors = ListedColormap(['red', 'blue', 'green', 'orange', 'purple', 'yellow'])
```

```python

# Plot data set pelatihan
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=colors, edgecolor='k')
plt.title('Visualising the Training set result')
plt.xlabel('RI')
plt.ylabel('Na')

# Tampilkan plot
plt.show()
```

```python
# Plot data set pengujian
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=colors, edgecolor='k')
plt.title('Visualising the Test set result')
plt.xlabel('RI')
plt.ylabel('Na')

# Tampilkan plot
plt.show()
```
